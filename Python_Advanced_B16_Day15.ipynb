```python
import numpy as np
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [1,4,9]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.title('Sales information')
plt.xlabel('sales')
plt.ylabel('revenue')
plt.plot(x,y)
plt.show()
```

```python
z = [10,5,0]
```

```python
plt.title('Sales information')
plt.xlabel('sales')
plt.ylabel('revenue')
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
plt.title('Sales info representation')
plt.xlabel('sales')
plt.ylabel('revenue')
plt.legend(["This is x&y","This is x&z"])
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```
